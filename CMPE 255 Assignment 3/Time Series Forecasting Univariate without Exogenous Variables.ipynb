Univariate time series forecasting without exogenous variables focuses on predicting future values based solely on past values of the series

```python
from pycaret.datasets import get_data
from pycaret.time_series import TSForecastingExperiment
e=TSForecastingExperiment()
```

```python
from pycaret.datasets import get_data
data=get_data('airline')
```

```python
e.setup(data=data,fh=10)
```

```python
best_model = e.compare_models()
```

```python
e.plot_model(best_model)
```
